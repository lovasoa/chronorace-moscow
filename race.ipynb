# Analyzing [chronotrack](https://www.chronotrack.com/) Race results
Data dowloaded from the API at https://results.chronotrack.com/m/ctlive/#29755/race/73970/786309/149013
using the following command :

```sh
seq 71 | parallel --jobs 10 wget -O "results/{}.json" "https://results.chronotrack.com/mobile/results/results?eventID=29755\&raceID=73970\&page={}\&limit=100"
```

## Loading the data
Load all the files, parse the json inside, read the array `results` inside each json object, and concatenate the results.

All that in just one line because python is so cool.

````python
import json, glob
````

````python
people = sum((json.load(open(f))["results"] for f in glob.iglob("results/*json")), [])
````

````python
people[0]
````

## Creating a Pandas object

````python
import pandas as pd
import numpy as np
import datetime
````

````python
raw_df = pd.DataFrame(people)
raw_df.sample(5)
````

````python
origins = raw_df["hometown"].str.split(", ", 1, expand=True)
males = raw_df["primary_bracket"].str.startswith("M")
origins[0] = origins[0].str.capitalize()
origins[1] = origins[1].str.capitalize()
````

````python
# columns=["name","hometown", "rank", "time", "pace", "primary_bracket"]
df = pd.concat((
    raw_df["rank"].astype("uint32"),
    raw_df["name"],
    origins.rename(columns=({0:"city",1:"country"})),
    pd.DataFrame({"is_male": males}),
    pd.to_timedelta(raw_df["time"]) / pd.Timedelta(seconds=1),
    pd.to_timedelta("00:" + raw_df["pace"]) / pd.Timedelta(seconds=1)
), axis=1)
df.set_index(["rank"], inplace=True)
df.sort_index(inplace=True)
df.head()
````

````python
df.dtypes
````

## Saving the results

````python
df.to_excel("race_results.xlsx")
````

## Plotting interesting values

````python
%matplotlib inline
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams["figure.figsize"]=[16,9]
````

````python
seconds_formatter = matplotlib.ticker.FuncFormatter(lambda x,pos: str(datetime.timedelta(seconds=round(x))))
percent_formatter = matplotlib.ticker.FuncFormatter(lambda y,pos: '{:.0%}'.format(y))
````

### Countries

````python
df.groupby("country").count() \
    .sort_values("name", ascending=False)["name"] \
    .plot(kind="bar", logy=True, color="darkblue")
````

````python
## Percentage of people from somewhere else
pd.Series(
   [sum(df["city"]=="Москва"), sum(df["city"]!="Москва")],
    index=["Moscow", "Other"]).plot.pie(figsize=(5,5), colors=["#aa88ee","#88eeaa"])
````

### Total time

````python
vals, bins = np.histogram(df["time"], bins=60)
fig = df[df["is_male"]]["time"].hist(bins=bins, color="#5555ff",alpha=.5)
df[~df["is_male"]]["time"].hist(bins=60, color="#ff9999", alpha=.5)
fig.set_xticks(bins[::5])
fig.xaxis.set_major_formatter(seconds_formatter)
fig.set_title("Histogram of finish time")
````

````python
#pace
vals, bins = np.histogram(df["pace"], bins=60)
fig = df[df["is_male"]]["pace"].hist(bins=bins, color="#5555ff", alpha=.6)
df[~df["is_male"]]["pace"].hist(bins=50, color="#ff9999", alpha=.6)
fig.set_xticks(bins[::5])
fig.set_xlabel("pace time (min/km)")
fig.set_ylabel("number of people")
fig.xaxis.set_major_formatter(seconds_formatter)
fig.set_title("Mean speed histogram")
````

````python
grouped = df.groupby("city")
data = grouped.median().sort_values("time")["time"][grouped.count()["time"] > 10][:10]
error = grouped.std()["time"][data.index]
fig = data.plot(kind="bar", yerr=error, title="Median time for the top-10 cities (with >10 participants)", color="#aa4455")
fig.yaxis.set_major_formatter(seconds_formatter)
````

````python
grouped = df.groupby("country")
data = grouped.median().sort_values("time")["time"][grouped.count()["time"] >= 5][:10]
error = grouped.std()["time"][data.index]
fig = data.plot(kind="bar", yerr=error, title="Median time for the top countries (with at least 5 participants)", color="#44aa55")
fig.yaxis.set_major_formatter(seconds_formatter)
````

````python
fig = df["time"].plot.line(figsize=(15,6), title="Total time = f(rank)", color="green", label="All")
plt.plot(df["time"][df["is_male"]].values, color="blue", label="Men")
plt.plot(df["time"][~df["is_male"]].values, color="#ee11aa", label="Women")
fig.set_ylabel("time (s)")
fig.set_xlabel("rank")
plt.legend()
fig.yaxis.set_major_formatter(seconds_formatter)
fig.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=10*60))
fig.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=400))
````

````python
def index_percent(old_series):
    percents = np.linspace(0,1,len(old_series))
    return pd.Series(percents, index=old_series)

fig = index_percent(df.time).plot(label="All", linewidth=2)
index_percent(df.time[df.is_male]).plot(label="Male")
index_percent(df.time[~df.is_male]).plot(label="Female")
fig.set_xlabel("time")
fig.set_ylabel("Position in the group")
plt.legend()
fig.xaxis.set_major_formatter(seconds_formatter)
fig.yaxis.set_major_formatter(percent_formatter)
fig.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=10*60))
fig.xaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(base=1*60))
fig.yaxis.set_major_locator(matplotlib.ticker.MultipleLocator(base=0.05))
fig.yaxis.set_minor_locator(matplotlib.ticker.MultipleLocator(base=0.01))
fig.patch.set_visible(False)
plt.grid(True, 'both', linewidth=.15, linestyle="-")
````
